```python
plt.figure(figsize=(10, 10))
plt.plot(x, f(x))
plt.grid(True)
plt.xlabel('График функции f(x) = (lg(x))^2 + (3/4)*lg(x) - 1/4')
```

```python
# [0.01, 0.5], [1.5, 2]
def bisec(eps, rootrange, f):
    a, b = rootrange[0], rootrange[1]
    x = (a + b)/2
    while b - a >= 2 * eps:
        if f(x) * f(a) < 0:
            b = x
        elif f(x) * f(b) < 0:
            a = x
        x = (a + b)/2
    return x
```

```python
root_bisec1 = bisec(1e-10, (0.01, 0.5), f)
root_bisec1
```

```python
root_bisec2 = bisec(1e-10, (1.5, 2), f)
root_bisec2
```

## Для функции g(x)

```python
x = np.arange(0.01, 3, 0.01)
plt.figure(figsize=(10, 10))
plt.plot(x, g(x))
plt.grid(True)
plt.xlabel('График функции g(x) = (lg(x))^2 + 2*lg(x) + 1 на отрезке [0; 3]')
```

Найдем корень:
$$\lg^2{x} + 2\lg{x} + 1 = 0$$
$$(\lg{x} + 1)^2 = 0$$
$$\lg{x} = -1$$
$$x = \frac{1}{10}$$

Найдем производную:
$$g(x) = \lg^2(x) + 2\lg(x) + 1$$

$$\frac{dg}{dx} = \frac{2\lg{x}}{x\ln{10}} + \frac{2}{x\ln{10}}$$
Решим уравнение:

$$\frac{2\lg{x}}{x\ln{10}} + \frac{2}{x\ln{10}} = 0$$

$$\frac{2\lg{x} + 2}{x\ln{10}} = 0$$

$$2\lg{x} + 2 = 0$$

$$2\lg{x} = -2$$

$$\lg{x} = -1$$

$$x = \frac{1}{10}$$

Значит, точка $x = 0.1$ является точкой экстремума. По графику видно, что на промежутке $[0.01; 3]$ эта точка является глобальным минимумом функции $g(x)$. Докажем это:

* Рассмотрим знакоопределенность производной на полуинтервалах $[0.01; 0.1)$ и $(0.1; 3.0]$:

$$\frac{dg}{dx}(0.02) = -30.3559$$

$$\frac{dg}{dx}(0.2) = 1.30736$$

* Как видим, на полуинтервале $[0.01; 0.1)$ производная функции $g(x)$ отрицательно определена. Значит, сама функция $g$ на этом интервале убывает;
* Как видим, на полуинтервале $(0.1; 3.0]$ производная функции $g(x)$ положительно определена. Значит, сама функция $g$ на этом интервале возрастает;
* Значит, точка $x = 0.1$ является точкой минимума. Так как на расматриваемом промежутке $[0.01; 3.0]$ не имеется больше точек экстремума, делаем вывод, что данная точка -- глобальный минимум функции на отрезке $[0.01; 3.0]$;

Так как точка $x = 0.1, g(x) = 0$ -- глобальный минимум функции на отрезке $[0.01; 3.0]$, то делаем вывод, что сама функция на этом отрезке положительно определена. А это значит, что функция на этом отрезке не меняет знак. Значит, применение метода бисекций для нахождения корня невозможно.

# Задача 2.2

```python
def f1(x):
    return (x + 1) * np.exp(x + 1) - x - 2
x = np.arange(-4, 2, 0.1)
```

```python
plt.figure(figsize=(10, 10))
plt.plot(x, f1(x))
plt.grid(True)
plt.xlabel('График функции f(x) = (x + 1)*Exp(x + 1) - x -2 на отрезке [-4; 2]')
```

Графически определим отрезки локализации корней:
* $[-3, -1]$
* $[-1, 0]$

Отрезок локализации наибольшего по модулю корня:
* $[-3, -1]$

### Найдем корень методом бисекций

```python
def bisec_itercount(eps, rootrange, f):
    a, b = rootrange[0], rootrange[1]
    x = (a + b)/2
    iters = 0
    while b - a >= 2 * eps:
        if f(x) * f(a) < 0:
            b = x
        elif f(x) * f(b) < 0:
            a = x
        x = (a + b)/2
        iters += 1
    return (x, iters)
```

```python
root_bisec_tk2 = bisec_itercount(1e-6, (-3, -1), f1)
print("Корень: {0};\nНайден за {1} итераций".format(np.round(root_bisec_tk2[0], 6), root_bisec_tk2[1]))
```

### Реализуем метод Ньютона и найдем корень

```python
def newton_itercount(eps, x0, f, f_deriv):
    x_prev = x0
    x_cur = x_prev - f(x_prev)/f_deriv(x_prev)
    
    iters = 1
    while np.abs(x_cur - x_prev) >= eps:
        x_prev = x_cur
        x_cur = x_cur - f(x_cur)/f_deriv(x_cur)
        
        iters += 1
    return (x_cur, iters)
```

```python
def f1_deriv(x):
    return np.exp(x + 1) * (x + 2) - 1
```

```python
newton_root_tk2 = newton_itercount(1e-6, -2, f1, f1_deriv)
print("Корень: {0};\nНайден за {1} итераций".format(np.round(newton_root_tk2[0], 6), newton_root_tk2[1]))
```

# Задача 2.3

$f(x) = (x - 1)*Sh(x + 1) - x$

```python
def f3(x):
    return (x - 1) * np.sinh(x + 1) - x
```

Локализуем корни:

```python
x = np.arange(-2, 2, 0.01)
plt.figure(figsize=(10, 10))
plt.plot(x, f3(x))
plt.grid(True)
plt.xlabel('График функции f(x) = (x - 1) * Sh(x + 1) - x на отрезке [-2; 2]')
```

## Пункт а)

Получаем отрезки локализации корней:
* $[-1; -0.5]$
* $[1;1.5]$

Приведем уравнение $f(x) = 0$ к виду $x = x - \alpha f(x)$ для корня на отрезке локализации $[-1; -0.5]$:

$\alpha = \frac{2}{M + m}$, 

$0 < m \le |f'(x)| \le M$:

* Вычислим $f'(x)$:
$$f'(x) = Sh(x + 1) + (x - 1)Ch(x + 1) - 1$$
$$f'(-1) = -3$$
$$f'(-0.5) = -2.17034$$

* Значит, $m = 2.17034; M=3$;

* Значит, $a = \frac{2}{3 + 2.17034} = \frac{2}{5.17034} = 0.38682$

* Выберем знак $a$ таким образом, чтобы он соответствовал знаку производной функции $f$ в окрестности корня: $a = -0.38682$

* Значит, получим в итоге уравнение: 
$$x = x+0.38682\dot{}((x - 1)\dot{}Sh(x + 1) - x)$$

```python
def simple_iter_var1(x0, l, f, q, eps):
    x_prev = x0
    x_cur = x_prev - l * f(x_prev)
    
    iters = 1
    while np.abs(x_cur - x_prev) >= ((1 - q)/q)*eps:
        x_prev = x_cur
        x_cur = x_prev - l * f(x_prev)
        
        iters += 1
    return (x_cur, iters)
```

```python
alpha = -0.38
m, M = 2.17034, 3
q = (M - m)/(M + m)
eps = 1e-5

def func(x):
    return (x - 1) * np.sinh(x + 1) - x
```

```python
res_simpleiter_pt_a_x1 = simple_iter_var1(-0.7, alpha, func, q, eps)
```

```python
print("Корень x1 = {0}\nПолучено за {1} итераций".format(np.round(res_simpleiter_pt_a_x1[0], 5), res_simpleiter_pt_a_x1[1]))
```

* Выберем знак $a$ таким образом, чтобы он соответствовал знаку производной функции $f$ в окрестности корня: $a = 0.18616$
